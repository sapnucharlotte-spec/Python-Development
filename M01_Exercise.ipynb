## Part 4 — print() and f-Strings

```python
name  = "Ana Reyes"
age   = 19
gpa   = 1.25

# Basic f-string
print(f"Name: {name}")
print(f"Age: {age}")
print(f"GPA: {gpa:.2f}")

# Expression inside {}
print(f"Birth year (approx): {2025 - age}")
```
