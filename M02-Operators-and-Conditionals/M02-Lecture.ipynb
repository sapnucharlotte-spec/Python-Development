# Part 1 — Operators

An **operator** is a symbol that performs an operation on one or more values.

You already know some from math: `+`, `-`, `*`, `/`. Python has more.

There are three groups you need to know:
- **Arithmetic** — do math
- **Comparison** — compare two values, always produces `True` or `False`
- **Logical** — combine True/False values

---
## Arithmetic Operators

| Operator | Name | Example | Result |
|----------|------|---------|--------|
| `+` | Addition | `5 + 3` | `8` |
| `-` | Subtraction | `5 - 3` | `2` |
| `*` | Multiplication | `5 * 3` | `15` |
| `/` | Division | `5 / 2` | `2.5` (always float) |
| `//` | Floor division | `5 // 2` | `2` (drops decimal) |
| `%` | Modulo (remainder) | `5 % 2` | `1` |
| `**` | Exponent | `2 ** 8` | `256` |

````python
# Run this and read each result
units     = 21
fee       = 500
passed    = 18
failed    = units - passed

total_fee = units * fee
average   = total_fee / units

print(f"Units enrolled : {units}")
print(f"Units failed   : {failed}")
print(f"Total fee      : {total_fee:,}")
print(f"Fee per unit   : {average:.2f}")
````

The `%` (modulo) operator is the one most beginners ignore — but it is surprisingly useful.

It gives you the **remainder** after division:
```
10 % 3 = 1   because   10 = (3 × 3) + 1
15 % 5 = 0   because   15 = (5 × 3) + 0
 7 % 2 = 1   because    7 = (2 × 3) + 1
```

The most common use: **checking if a number is even or odd**.
- If `number % 2 == 0` → even
- If `number % 2 == 1` → odd

## Comparison Operators

Comparison operators **compare two values** and always return `True` or `False`.

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `==` | Equal to | `5 == 5` | `True` |
| `!=` | Not equal to | `5 != 3` | `True` |
| `>` | Greater than | `5 > 3` | `True` |
| `<` | Less than | `5 < 3` | `False` |
| `>=` | Greater than or equal | `5 >= 5` | `True` |
| `<=` | Less than or equal | `3 <= 5` | `True` |

Remember from M01: `=` is **assignment** (store a value). `==` is **comparison** (are these equal?). Mixing them up is one of the most common mistakes in Python.

````python
grade = 88
passing = 75

print(grade >= passing)    # Is the student passing?
print(grade == 100)        # Is it a perfect score?
print(grade != passing)    # Is the grade different from the passing mark?
print(grade > 90)          # Is it above 90?
````

## Logical Operators

Logical operators let you **combine multiple conditions** into one.

| Operator | Meaning | Result |
|----------|---------|--------|
| `and` | Both must be True | `True and True` → `True` |
| `or` | At least one must be True | `True or False` → `True` |
| `not` | Flips True to False and vice versa | `not True` → `False` |

Think of it in plain English:
- **and** — "You pass if your grade is above 75 **and** you have no absences"
- **or** — "You get a discount if you are a senior **or** a PWD"
- **not** — "You cannot enroll if you are **not** yet cleared"

````python
grade    = 80
absences = 2
cleared  = True

# and — both conditions must be True
print(grade >= 75 and absences <= 3)   # passing grade AND acceptable absences?

# or — at least one must be True
is_senior = False
is_pwd    = True
print(is_senior or is_pwd)             # qualifies for discount?

# not — flips the value
print(not cleared)                     # is the student NOT yet cleared?
print(not (grade < 75))                # is the grade NOT failing?
````

# Part 2 — Conditionals

Operators let you **ask questions**. Conditionals let your program **act on the answers**.

The basic structure:

```python
if condition:
    # runs only if condition is True
elif another_condition:
    # runs if the first was False but this one is True
else:
    # runs if none of the above were True
```

Two things that trip up beginners:
1. The **colon** `:` at the end of `if`, `elif`, `else` — it is required
2. **Indentation** — the code inside the block must be indented (4 spaces or 1 tab). Python uses indentation to know what belongs inside the block.

````python
# Basic if statement
grade = 88

if grade >= 75:
    print("Passed")
````

````python
# if / else
grade = 65

if grade >= 75:
    print("Passed")
else:
    print("Failed")
````

````python
# if / elif / else — multiple conditions
# Philippine grading system
grade = 88

if grade >= 90:
    print("Excellent — 1.0")
elif grade >= 85:
    print("Very Good — 1.25 or 1.50")
elif grade >= 80:
    print("Good — 1.75")
elif grade >= 75:
    print("Satisfactory — 2.0 to 3.0")
else:
    print("Failed — 5.0")
````

Important: Python checks conditions **top to bottom** and stops at the first one that is `True`. Once a match is found, the rest are skipped.

This means order matters — always put more specific conditions before more general ones.

````python
# PREDICT
# What does this print? Think carefully about the order.

# My guess: ___

score = 95

if score >= 70:
    print("Passed")
elif score >= 90:
    print("Excellent")
else:
    print("Failed")
````

# Part 3 — Nested Conditionals

You can put an `if` statement inside another `if` statement. This is called **nesting**.

Use it when the second decision only makes sense after the first condition is already met.

```python
if outer_condition:
    if inner_condition:
        # both are True
    else:
        # outer is True, inner is False
else:
    # outer is False
```

Real example: You can only check a student's GPA category **after** confirming they are enrolled.

````python
is_enrolled = True
gpa         = 1.50
balance     = 0

if is_enrolled:
    if balance > 0:
        print("You have an outstanding balance. Please settle before proceeding.")
    else:
        if gpa <= 1.75:
            print("You are on the Dean's List. Congratulations!")
        else:
            print("You are a regular student in good standing.")
else:
    print("You are not currently enrolled.")
````

Nested conditions work but can get hard to read quickly. A good rule of thumb: if you are more than 3 levels deep, consider using `and`/`or` to flatten it out.

For example, the above can be partially simplified:

````python
# Same logic, flatter structure using 'and'
is_enrolled = True
gpa         = 1.50
balance     = 0

if is_enrolled and balance == 0 and gpa <= 1.75:
    print("Dean's List")
elif is_enrolled and balance == 0:
    print("Regular student in good standing")
elif is_enrolled and balance > 0:
    print("Outstanding balance — please settle")
else:
    print("Not enrolled")
````

x < y    # less than

# --- LOGICAL ---
x and y  # True only if both are True
x or y   # True if at least one is True
not x    # flips True to False and vice versa

# --- CONDITIONALS ---
if condition:
    # runs if True
elif another_condition:
    # runs if previous was False and this is True
else:
    # runs if none matched
