## Part 2 — Data Types

```python
# The four basic types
print(type(42))        # int
print(type(3.14))      # float
print(type("Hello"))   # str
print(type(True))      # bool
```

```python
# Division behavior — important to know
print(10 / 3)    # always float
print(10 // 3)   # floor — drops decimal
print(10 % 3)    # remainder
```

## Part 4 — print() and f-Strings

```python
name  = "Ana Reyes"
age   = 19
gpa   = 1.25

# Basic f-string
print(f"Name: {name}")
print(f"Age: {age}")
print(f"GPA: {gpa:.2f}")

# Expression inside {}
print(f"Birth year (approx): {2025 - age}")
```

```python
# TRY
# Using ONLY f-strings, print the output below exactly.
# Do not hardcode values — use the variables provided.
#
# Expected output:
# ==============================
# Name    : Maria Santos
# Course  : BSIT
# GPA     : 1.75
# ==============================

name   = "Maria Santos"
course = "BSIT"
gpa    = 1.75

# Your code here
print("=" * 30)
print(f"Name    : {name}")
print(f"Course  : {course}")
print(f"GPA     : {gpa}")
print("=" * 30)
```

```python
# EXPLORE
# f-strings support alignment formatting.
# Run this and study the output, then try making your own table below.

print(f"{'Name':<20} {'GPA':>6}")
print(f"{'Ana Reyes':<20} {1.25:>6.2f}")
print(f"{'Jose Santos':<20} {1.50:>6.2f}")
print(f"{'Maria dela Cruz':<20} {1.75:>6.2f}")
print(f"{'Lian San Diego':<20} {1:>6.2f}")
print(f"{'Jenny Bugay':<20} {1.85:>6.2f}")

# Try adding two more rows using your classmates' names (made up is fine)

```

## Part 5 — input()

```python
# input() always returns a string
name = input("What is your name? ")
print(f"Hello, {name}!")
print(f"Type: {type(name)}")
```

```python
# TRY
# Write a program that:
# 1. Asks for the user's name
# 2. Asks for their age (as an integer)
# 3. Asks for their GPA (as a float)
# 4. Prints a formatted summary using f-strings
#
# Sample output:
# ----------------------------
# Name : Juan dela Cruz
# Age  : 20
# GPA  : 1.75
# ----------------------------

name  = input("What is your name? ")
age   = int(input("What is your age? "))
gpa   = float(input("What is your GPA? "))

print("----------------------------")
print(f"Name : {name}")
print(f"Age  : {age}")
print(f"GPA  : {gpa}")
print("----------------------------")
```
