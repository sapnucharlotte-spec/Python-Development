# Part 1 — Variables

## What is a variable?

Imagine you're filling out a form. You write your name in a box labeled **"Name"**, your age in a box labeled **"Age"**, and so on. Each labeled box holds one piece of information.

A **variable** works the same way — it's a **named container** that holds a value in your program's memory.

```
┌──────────────┐       ┌──────────────┐       ┌──────────────┐
│   student    │       │     age      │       │     gpa      │
│  "Maria"     │       │      20      │       │    1.75      │
└──────────────┘       └──────────────┘       └──────────────┘
```

## Creating a variable

In Python, you create a variable by writing:

```
variable_name = value
```

The `=` sign here does **not** mean "equals" like in math. It means **"store this value into this variable"**. We call this **assignment**.

Run the cell below to see variables in action:

````python
# Creating variables
student_name = "Maria Santos"
age          = 20
gpa          = 1.75
is_enrolled  = True

# Displaying them
print(student_name)
print(age)
print(gpa)
print(is_enrolled)
````

Notice we stored different kinds of values — text, a number, a decimal, and a true/false. Python handles all of them.

---

## Naming Rules

Not every name is a valid variable name. Python has strict rules:

| Rule | Valid | Invalid |
|------|---------|----------|
| Must start with a letter or `_` | `name`, `_score` | `1name`, `@score` |
| Can contain letters, numbers, `_` | `student_1` | `student-1`, `student 1` |
| Case-sensitive | `name` ≠ `Name` ≠ `NAME` | — |
| Cannot be a Python keyword | `score` | `if`, `for`, `print`, `True` |

**Convention:** Python programmers use `snake_case` — all lowercase with underscores between words.

```python
# Good naming
first_name = "Juan"
total_score = 95
is_passing = True

# Bad naming (works but not recommended)
FirstName = "Juan"
TOTALSCORE = 95
x = True
```

````python
# TRY IT
# Variables can be updated (reassigned) at any time
# Watch what happens to the value of 'score'

score = 75
print("Score before:", score)

score = 90          # reassigning — the old value 75 is replaced
print("Score after:", score)

score = score + 5   # using the current value to compute a new one
print("Score + 5:", score)
````

# Part 2 — Data Types

## What is a data type?

Every value in Python has a **type** — it tells Python what kind of data it is and what operations make sense for it.

Think of it this way: you can multiply two numbers, but you can't multiply two names. The data type is what enforces these rules.

The four basic types you'll use constantly:

| Type | Full Name | What it holds | Example |
|------|-----------|---------------|---------|
| `int` | Integer | Whole numbers (no decimal) | `21`, `-5`, `0`, `1000` |
| `float` | Floating-point | Numbers with decimals | `3.14`, `1.75`, `-0.5` |
| `str` | String | Text — any characters in quotes | `"Hello"`, `'BSIT'`, `"123"` |
| `bool` | Boolean | Only two possible values | `True` or `False` |

> ⚠️ **Important:** `"123"` is a **string**, not a number! Quotes make it text. `123` without quotes is an **integer**.

````python
# You can check the type of any value using type()

print(type(21))        # int
print(type(3.14))      # float
print(type("Hello"))   # str
print(type(True))      # bool
print(type("123"))     # str  ← notice: quotes make it a string!
print(type(123))       # int  ← no quotes = number
````

---
## Integers and Floats

Both `int` and `float` are numbers, but they behave slightly differently.

**When to use `int`:** counting things (number of students, years, items)  
**When to use `float`:** measuring things (GPA, price, temperature, percentage)

One important thing to know — **division in Python always returns a float**, even if the result is a whole number:

````python
# Regular division — always returns float
print(10 / 2)      # 5.0  ← float, not 5!
print(7 / 2)       # 3.5

# Floor division — returns int (cuts off the decimal)
print(7 // 2)      # 3   ← drops the .5

# Modulo — returns the remainder
print(7 % 2)       # 1   ← because 7 = (2 × 3) + 1

# Exponent
print(2 ** 8)      # 256
````

---
## Strings

A **string** is any sequence of characters inside quotes. It can be a name, a sentence, a single letter, even a number written in quotes.

You can use single `'...'` or double `"..."` quotes — both work. Just be consistent.

````python
# Different ways to write strings
greeting = "Hello, world!"
name     = 'Juan dela Cruz'
number   = "42"              # this is a STRING, not a number

# Combining strings (concatenation) using +
first = "Juan"
last  = "dela Cruz"
full  = first + " " + last   # space added manually
print(full)

# Repeating strings using *
print("-" * 30)              # prints 30 dashes
print("Ha" * 5)              # prints HaHaHaHaHa

# Length of a string
print(len("Python"))         # 6
print(len(full))             # counts all characters including space
````

---
## Booleans

A **boolean** has only two possible values: `True` or `False`. (Capital T and F — this matters!)

Booleans are the result of any **comparison** in Python. You'll use them heavily when you get to `if` statements in Module 3.

````python
# Direct boolean values
is_enrolled = True
has_failed  = False

print(is_enrolled)
print(has_failed)

# Comparisons produce booleans
grade = 88
print(grade >= 75)    # True  — is 88 greater than or equal to 75?
print(grade == 100)   # False — is 88 equal to 100?
print(grade != 100)   # True  — is 88 NOT equal to 100?

# type() confirms
print(type(True))     # <class 'bool'>
````

---
## Type Conversion

Sometimes you have a value of one type but need it in another. Python gives you built-in functions to convert:

| Function | Converts to | Example |
|----------|-------------|----------|
| `int()` | Integer | `int("42")` → `42` |
| `float()` | Float | `float("3.14")` → `3.14` |
| `str()` | String | `str(100)` → `"100"` |
| `bool()` | Boolean | `bool(0)` → `False` |

This comes up **constantly** in real programs — especially when reading user input (which is always a string).

````python
# String → int
age_text = "20"
age_num  = int(age_text)
print(age_num + 5)      # 25 — now we can do math!

# String → float
gpa_text = "1.75"
gpa_num  = float(gpa_text)
print(gpa_num + 0.25)   # 2.0

# int → string (needed when combining text + number)
year = 2
print("Year " + str(year) + " BSIT")

# float → int (WARNING: it TRUNCATES, does NOT round)
print(int(3.99))        # 3  ← not 4!
print(int(3.01))        # 3
````

# Part 3 — Output with `print()`

You've been using `print()` already. Now let's understand it properly and learn how to format output cleanly.

## Basic `print()`

`print()` can take one or multiple values separated by commas. By default it puts a space between them and a newline at the end.

````python
# One value
print("Hello!")

# Multiple values — separated by space automatically
print("Name:", "Maria", "Age:", 20)

# Custom separator using sep=
print("2025", "06", "15", sep="-")     # 2025-06-15
print("Juan", "dela Cruz", sep=", ")   # Juan, dela Cruz

# Prevent newline at end using end=
print("Loading", end="")
print("...", end="")
print(" Done!")

# Print a blank line
print()
print("After the blank line")
````

---
## f-Strings — The Modern Way to Format Output

An **f-string** (formatted string literal) lets you embed variables and expressions directly inside a string. Just put `f` before the opening quote and wrap variables in `{}`.

This is the **recommended** approach in modern Python (3.6+). You will use this almost every time you print something.

```python
# Old way (messy)
print("Hello, " + name + "! You are " + str(age) + " years old.")

# f-string way (recommended)
print(f"Hello, {name}! You are {age} years old.")
```

````python
name  = "Ana Reyes"
age   = 19
grade = 1.25

# Basic f-string
print(f"Student: {name}")
print(f"Age: {age}")

# Expressions inside {}
print(f"Birth year (approx): {2025 - age}")
print(f"Age next year: {age + 1}")

# Controlling decimal places with :.2f
print(f"Grade: {grade:.2f}")      # always 2 decimal places
print(f"Grade: {grade:.4f}")      # 4 decimal places

# Padding and alignment (useful for tables)
print(f"{'Name':<20} {'Grade':>10}")
print(f"{'Ana Reyes':<20} {1.25:>10.2f}")
print(f"{'Jose Santos':<20} {1.50:>10.2f}")
print(f"{'Maria dela Cruz':<20} {1.75:>10.2f}")
````

# Part 4 — Input with `input()`

## What is `input()`?

So far our programs have had **fixed values** — we write `name = "Maria"` directly in the code. But real programs need to work with **data from the user**.

The `input()` function does this:
1. Displays a **prompt message** to the user
2. **Pauses** and waits for the user to type something and press Enter
3. **Returns** whatever the user typed as a **string**

```python
variable = input("Prompt message here: ")
```

> ⚠️ **CRITICAL RULE:** `input()` **ALWAYS returns a string** — even if the user types a number. You must convert it if you need to do math.

````python
# Basic input
# Run this cell — it will ask you to type your name
name = input("What is your name? ")
print(f"Welcome, {name}!")
print(f"Type of name: {type(name)}")
````

````python
# Getting a number — must convert!
age_input = input("Enter your age: ")
print(f"Type before conversion: {type(age_input)}")  # str!

age = int(age_input)  # convert str → int
print(f"Type after conversion:  {type(age)}")        # int!

print(f"In 10 years you will be {age + 10} years old.")
````

````python
# Shorthand — convert directly on the same line (most common pattern)
age    = int(input("Enter your age: "))
height = float(input("Enter your height in meters (e.g. 1.65): "))

print(f"\n--- Your Info ---")
print(f"Age:    {age} years old")
print(f"Height: {height:.2f} m")
````

````python
# A complete mini-program using input + f-strings
# Run this and enter your info

print("=" * 35)
print("   STUDENT INFORMATION FORM")
print("=" * 35)

name     = input("Full name: ")
age      = int(input("Age: "))
course   = input("Course: ")
units    = int(input("Number of units enrolled: "))
fee      = float(input("Fee per unit (₱): "))

total = units * fee

print()
print("=" * 35)
print("   SUMMARY")
print("=" * 35)
print(f"  Name    : {name}")
print(f"  Age     : {age}")
print(f"  Course  : {course}")
print(f"  Units   : {units}")
print(f"  Total   : ₱{total:,.2f}")
print("=" * 35)
````

# Part 5 — Common Mistakes

These are the errors beginners make most often in this module. Read through them carefully — you will definitely encounter some of these.

---

### Mistake 1 — Forgetting to convert `input()`

````python
# Run this and type a number — watch what happens
x = input("Enter a number: ")
print(x + x)          # Does NOT add! It JOINS strings instead

# Fix:
x = int(input("Enter a number: "))
print(x + x)          # Now it adds correctly
````

### Mistake 2 — Mixing `str` and `int` in `print()`

````python
age = 20

# This will ERROR — cannot concatenate str and int
# print("I am " + age + " years old.")    # uncomment to see error (because age is an int, not a str)

# Fix 1 — use str() to convert
print("I am " + str(age) + " years old.")

# Fix 2 — use f-string (easier)
print(f"I am {age} years old.")

# Fix 3 — use comma in print()
print("I am", age, "years old.")
````

### Mistake 3 — Thinking `int()` rounds numbers

````python
# int() does NOT round — it TRUNCATES (cuts off decimal)
print(int(3.9))    # 3  ← NOT 4!
print(int(9.99))   # 9  ← NOT 10!
print(int(-2.7))   # -2 ← NOT -3!

# If you want to round, use round() instead
print(round(3.9))  # 4  ✅
print(round(3.5))  # 4  ✅
````

### Mistake 4 — Using `=` instead of `==` for comparison

````python
score = 90

# = means ASSIGNMENT (store a value)
# == means COMPARISON (check if equal)

print(score == 90)   # True  — asking: is score equal to 90?
print(score == 80)   # False — asking: is score equal to 80?
print(score != 80)   # True  — asking: is score NOT equal to 80?
````
